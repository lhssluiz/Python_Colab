### Criando e usando classes
The *class* will reproduce the major behaviour of a object and the object is a instance of it. 

all the time that we instance dog, we'll be representing it. And we can make specific instances of a dog[texto do link](https://)

```python
class Dog():
  '''It's a simple try to model a dog'''
  def __init__(self,name,age):
    self.name = name
    self.age = age
  
  def sit(self):
    '''Simulates a dog sitting in a answer to a command'''
    print(self.name.title() + " is now sitting.")

  def roll_over(self):
    '''Simulates a dog rolling over in a answer to a command'''
    print(self.name.title() + " rolled over!")

# Start by the definition of the class, with the reserved word class
# The convention is to use the first letter in capital format
# A function as a part of a class, it's called a method
# The method _init_() is a special method in Python that execute automatically
# always that a instance based in Dog is created
```

```python
'''my_dog is self in this case
The name and the age is given by the other arguments

init creates an instance (a space in the memory) to store
self, name and age. It hasn't a specific _return_, but it 
return the automatically the instance that represents this dog
'''
my_dog = Dog('willie',6)

print("My dog's name is " + my_dog.name.title() + ".")
print("My dog is " + str(my_dog.age) + " years old.")
```

```python
## Calling the methods inside the class
#
# Specify the name of the instance created, then the method separated by (.)
my_dog.sit()
my_dog.roll_over()
```

```python
your_dog = Dog('wilson',17)
print("My dog's name is " + my_dog.name.title() + ".")
print("My dog is " + str(my_dog.age) + " years old.")
print("\n")
print("My dog's name is " + your_dog.name.title() + ".")
print("My dog is " + str(your_dog.age) + " years old.")
```

###### Incrementando um valro de um atributo com um método

```python
# Adding update odometer

class Car():
  '''A simple sample to represent a car'''
  
  def __init__(self,make,model,year):
    '''Initiate the attributes that decribe a car'''
    self.make = make
    self.model = model
    self.year = year
    self.odometer_reading = 0

  def get_decriptive_name(self):
    '''Return a decriptive name, elegantly formatted'''
    long_name = str(self.year) + ' ' + self.make + ' ' + self.model
    return long_name.title()
  
  def read_odometer(self):
    '''Shows a sentence with the car's miles'''
    print("This car has " + str(self.odometer_reading) + " miles on it.")
  
  def update_odometer(self,mileage):
    '''Method to update odometer and rejects minor values than the current'''
    if mileage >= self.odometer_reading:
      self.odometer_reading = mileage
    else:
      print("You can't roll back an odometer!")
  
  def increment_odometer(self,miles):
    '''Sum que amount specified to the reading value'''
    self.odometer_reading += miles

  
```

```python
my_used_car = Car('subaru','outback',2013)
```

```python
print(my_used_car.get_decriptive_name())
my_used_car.update_odometer(23500)
my_used_car.read_odometer()
my_used_car.increment_odometer(100)
my_used_car.read_odometer()
```

```python
# 9.4 People attended (9.1)

class Restaurant:
  '''Class to deal with all restaurant need'''

  def __init__(self,name,description):
    '''Initialize with 3 variables, where 1 is default'''
    self.name = name
    self.description = description
    self.number_served = 0
  
  def describe_restaurant(self):
    '''Shows restaurant description'''
    print("The restarant " + self.name + " has this cuisine type " + self.description +"." )

  def open_restaurant(self):
    '''Shows if restaurant is open'''
    print("The restaurant " + self.name + " is open!")
  
  def set_number_served(self,clients_served):
    '''Sotre the served number'''
    self.number_served = clients_served
  
  def increment_number_served(self,new_clients):
    '''Increment the number of clients served'''
    self.number_served += new_clients
```

```python
restaurant = Restaurant('Roxy', 'Italian')

# restaurant.name
# restaurant.description
print("The number of people served here: " + str(restaurant.number_served))

restaurant.set_number_served(32)
print("The number of people served here: " + str(restaurant.number_served))
restaurant.increment_number_served(25)
print("The number of people served here: " + str(restaurant.number_served))
```

```python
#9.5 Login attempts (9.3)

class User:

  def __init__(self,first_name,last_name,id_info,log_attempts):
    self.name = first_name
    self.surname = last_name
    self.id = id_info
    self.login_attempts = log_attempts
  
  def describe_user(self):
    '''Describe the user informations'''
    print("This are all the informations, about me: ")
    print("\t-" + self.name)
    print("\t-" + self.surname)
    print("\t-" + str(self.id))
  
  def greet_user(self):
    '''Shows a greeting to the user'''
    print("Hello "+ self.name + " " + self.surname + " I'm glad to see you here!")
  
  def increment_login_attempts(self):
    '''Increment the login attempts of a user'''
    self.login_attempts += 1
  
  def reset_login_attempts(self):
    '''Back the login attempts to zero'''
    self.login_attempts = 0


user1 = User('Luiz','Sá',999,0)
user1.increment_login_attempts()
print(str(user1.login_attempts))
user1.increment_login_attempts()
user1.increment_login_attempts()
print(str(user1.login_attempts))
user1.reset_login_attempts()
print(str(user1.login_attempts))
```
