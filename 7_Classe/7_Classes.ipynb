### Criando e usando classes
The *class* will reproduce the major behaviour of a object and the object is a instance of it. 

all the time that we instance dog, we'll be representing it. And we can make specific instances of a dog[texto do link](https://)

````python
class Dog():
  '''It's a simple try to model a dog'''
  def __init__(self,name,age):
    self.name = name
    self.age = age
  
  def sit(self):
    '''Simulates a dog sitting in a answer to a command'''
    print(self.name.title() + " is now sitting.")

  def roll_over(self):
    '''Simulates a dog rolling over in a answer to a command'''
    print(self.name.title() + " rolled over!")

# Start by the definition of the class, with the reserved word class
# The convention is to use the first letter in capital format
# A function as a part of a class, it's called a method
# The method _init_() is a special method in Python that execute automatically
# always that a instance based in Dog is created
````

````python
'''my_dog is self in this case
The name and the age is given by the other arguments

init creates an instance (a space in the memory) to store
self, name and age. It hasn't a specific _return_, but it 
return the automatically the instance that represents this dog
'''
my_dog = Dog('willie',6)

print("My dog's name is " + my_dog.name.title() + ".")
print("My dog is " + str(my_dog.age) + " years old.")
````

## Herança
The heritage is useful to reuse another class. If you're writting a specialized form of a exsited class, you can use this heritage. When a class inherit from other, it assumes all the attibutes from the father class

Father Class > Child Class

````python
# v1 Creating a class ElectricCar as a child to car

class Car():
  '''A simple sample to represent a car'''
  
  def __init__(self,make,model,year):
    '''Initiate the attributes that decribe a car'''
    self.make = make
    self.model = model
    self.year = year
    self.odometer_reading = 0

  def get_decriptive_name(self):
    '''Return a decriptive name, elegantly formatted'''
    long_name = str(self.year) + ' ' + self.make + ' ' + self.model
    return long_name.title()
  
  def read_odometer(self):
    '''Shows a sentence with the car's miles'''
    print("This car has " + str(self.odometer_reading) + " miles on it.")
  
  def update_odometer(self,mileage):
    '''Method to update odometer and rejects minor values than the current'''
    if mileage >= self.odometer_reading:
      self.odometer_reading = mileage
    else:
      print("You can't roll back an odometer!")
  
  def increment_odometer(self,miles):
    '''Sum que amount specified to the reading value'''
    self.odometer_reading += miles

class ElectricCar(Car):
  '''Represents sprcific aspects of electric vehicles'''

  def __init__(self,make,model,year):
    '''Initialize the attributes of dad's class'''
    super().__init__(make, model,year)

my_tesla = ElectricCar('tesla','model s',2016)
print(my_tesla.get_decriptive_name())

# When to create a child class, the dad's class needs to be part of teh same file
# The name of the dad's class needs to be inserted in the definition of the
# child class
# The method __init__() accepts the informations needed to create a instance of Car
# The function super, is a special function that helps Python to create
# conections between the dad's class and the child, it passes all the attributes
# of the dad's class
# This name is because of the convention - that I'll use towards on - of the name
# superclass (parent) and subclass
````

### Sobrescrever métodos da classe pai
When you want to overwrite a method from the superclasse, because it doesn't fit to the subclass, you need to create in the subclass a method with the same name, then, when you'll instanciate it, Python will treat the method as it's defined in the subclass. Ex:

Super clas has a method called "fill_gas_tank()" it doesn't fir to electric cars, then you create a "fill_gas_tank()" in the subclass and the new behaviours to this specific usage

#### Instâncias como atributos
In case you're writting a class and realize the it's becoming too much big. Then, you can create another separated class and instanciate the new Class to the "previous" one to use the behaviours

````python
# Using battery as a attibute to ElectricCar

class Battery():
  '''A try to simply modelate a battery'''

  def __init__(self, battery_size=70):
    '''initialize the attributes of the battery'''
    self.battery_size = battery_size
  
  def describe_battery(self):
    '''Presents a sentence to shows the capacity of the battery'''
    print("This car has a " + str(self.battery_size) + "-kWh battery.")
  
  def get_range(self):
    '''Shows a sentence aobut the distance that 
    the car is capable to run with this battery level'''
    if self.battery_size == 70:
      range = 240
    elif self.battery_size == 85:
      range = 270
    message = "This car can go approximately " + str(range) 
    message += " miles on a full charge."
    print(message) 
    #A way to present two lines of text

class ElectricCar(Car):
  '''Represents specific aspects of electric vehicles
  Then initialize the attributes for electric cars'''

  def __init__(self,make,model,year):
    '''Initialize the attributes of dad's class'''
    super().__init__(make, model,year)
    self.battery = Battery()

my_tesla = ElectricCar('tesla','model s',2016)
print(my_tesla.get_decriptive_name())
my_tesla.battery.describe_battery()
my_tesla.battery.get_range()

# ElectricCar has now one instance of Battery
# Whrn we want to decribe the battery now we need:
# We use the instance 'my_tesla' that now, contains also
# a instance of ElectricCar, since it was created a instance
# then, we need to call the attribute Battery 'my_tesla.battery'
# and finally call the method that we want to use 
# 'my_tesla.battery.describe_battery'
#
# It's a good solution, because we don't need to insert all the 
# possible and new details to Electric car, however, we use it in Battery()
````

#### Armazenando várias classes em um módulo
You can store as much as classes will be necessary in only one module. But, the classes needs to be related to each other in some way

**You'll find the example in files my_electric.ipynb and car.py**

#### Imoprting more than one class
To do so, you just need to separate them by comma after import the file, ex:
from car import ElectricCar,Car

#### Importando um módulo inteiro
To do so, you only import the module, like:
```
import car
```


And during the code, use the (.) dot notation, ex:

```
my_beetle = car.Car(arguments)
```


As you can see, before we address 'car' as a module and then we call the class 'Car".

#### Importando todas as classes de um módulo
To import all the casses, you just need to add a asterisc to the notation, like this:
```
from name_module import*
```
**This option is not recommended, because it can generate confusion**. Import all the module it's better, because with this you use the dot notation

#### Importando um módulo em um módulo
If you find a situation where a class in one module (1st) depends of a class in other module (2nd). You'll import the class needed in the 1st module to the 2nd file in this example, to make this one that depends of the 1st to work

As example we could separate the class Car and ElectricCar, Battery in two different modules. If this would happen, since ElectricCar is a subclass of Car, we would need to import Car to the file with ElectricCar. After this the could would work properly.

In the case that it would use a third file, it would be needed to import both modules to make them work in the third one
