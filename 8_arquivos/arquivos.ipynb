Treating Python errors. using the block try-except

`````python
try:
  print(5/0)
except ZeroDivisionError:
  print("It's not possible to make this operation")
`````

`````python
# A simple calculator
print("Give me two numbers and I'll divide them.")
print("Enter 'q' to quit. ")
while True:
  first_number = input("Insert the 1st number: ")
  if first_number == 'q':
    break
  second_number = input("Insert the 2nd number: ")
  if second_number == 'q':
    break
  answer = int(first_number) / int(second_number)
  print(str(answer))
`````

`````python
# A simple calculator with error treatment

print("Give me two numbers and I'll divide them.")
print("Enter 'q' to quit. ")
while True:
  first_number = input("Insert the 1st number: ")
  if first_number == 'q':
    break
  second_number = input("Insert the 2nd number: ")
  if second_number == 'q':
    break
  try:
    answer = int(first_number) / int(second_number)
  except ZeroDivisionError:
    print("It's not possible to divide by zero")  
  else:
    print(str(answer))

# In this case, any code that depends of the success of "try"
# will be presented inside the else. This is the way to keep programs
# working with errors
`````

### Armazenando dados
It's possible to store data to use after. We'll make it here using JSON

##### Usando json.dump() e json.load()
json.dump() armazena os dados e json.load() carrega estes dados para serem usados

`json.dump(dado_p/_armazenar, objetvo_arquivo)`

`````python
import json

numbers = [2,3,5,7,11,13]

filename = '/content/drive/MyDrive/Colab_Notebooks/8_arquivos/numbers.json'
with open(filename, 'w') as f_obj:
  json.dump(numbers,f_obj)

with open(filename) as fi_obj:
  numbers = json.load(fi_obj)
print(numbers)
`````

#### Refatoração
Refactoring let's the code more clean, easy to read and to extend it

`````python
# Refactoring the code above
def greet_user():
  '''Present a greeting for a user
  save its data inside a json
  evaluate if the file exist'''
  filename = 'username.json'

  try:
   with open(file_path+filename) as file_obj:
    username = json.load(file_obj)
  except FileNotFoundError:
    username = input("Plese, say you name for us: ")
    with open(file_path+filename, 'w') as file_obj:
      json.dump(username, file_obj)
      print("We'll remember your name " + username + "!")
  else:
    print("Welcome back " + username + "!")
`````

`````python
# Refactoring 3.0
def get_stored_username():
  '''Get the username stored if it exist'''
  filename = 'username.json'
  try:
   with open(file_path+filename) as file_obj:
    username = json.load(file_obj)
  except FileNotFoundError:
    return None
  else:
    return username
# It's a good practice to return None instead of the expected result
# it makes easiest to test

def get_new_username():
  '''Asks for a new name'''
  username = input("What's your name? ")
  filename = 'username.json'
  with open(file_path+filename) as file_obj:
    json.dump(username, file_obj)

def greet_user():
  '''Print a greet to the user'''
  username = get_stored_username()
  if username:
    print("Welcome back " + username + "!")
  else:
    username = get_new_username()
    print("We'll remember your name in the next, " + username)
    
`````
